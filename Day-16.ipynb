```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("StudentsPerformance.csv") #load the data
df
```

```python
df.head() #First 5
# df.tail() Last 5
```

```python
sns.relplot(x='reading score',y='writing score',data=df,hue='gender') #Scatter plot-->relation b/w reading and writing scote and differentiating b/w male nd female
```

```python
sns.relplot(x='reading score',y='writing score',data=df,hue='gender',kind='line')  # changed pattern of scatter plot(kind=line)
```

```python
sns.catplot(x='race/ethnicity',y='math score',data=df,kind='bar')
```

```python
sns.catplot(x='race/ethnicity',y='math score',data=df,kind='violin')
plt.title('JECRC') #gives title
plt.xlabel("Group")
plt.ylabel("Math score")
# kind--> violin, strip, bar, swarn
```

```python
sns.catplot(x='race/ethnicity',y='math score',data=df,kind='strip',hue='gender',col='gender' ) # differentiating math score of diff groups and agin seperating them on the basis of their gender
plt.show() # removes unnecasry values
```
